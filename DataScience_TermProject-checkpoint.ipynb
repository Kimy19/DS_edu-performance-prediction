### imd_band: deprivation rate of small areas.
### fill missing data with each data's region's mean of imd_band value

### Encoding imd_band

```python
# Ordinal encoding 
imd_band_encode={
'0-10%':0,
'10-20':1,
'20-30%':2,
'30-40%':3,
'40-50%':4,
'50-60%':5,
'60-70%':6,
'70-80%':7,
'80-90%':8,
'90-100%':9}
studInfo['imd_band']=studInfo.imd_band.map(imd_band_encode)
print(studInfo['imd_band'].value_counts())
```

```python
region_values=studInfo['region'].unique()
print(region_values)
```

```python
# Fillna the nan value of imd_band with the average imd_band value of the 'region' to which nan data belongs.
for i in range(len(region_values)):
    imd_mean=studInfo[studInfo['region']==region_values[i]]['imd_band'].mean()
    studInfo.loc[studInfo['region']==region_values[i],'imd_band']=studInfo[studInfo['region']==region_values[i]]['imd_band'].fillna(imd_mean)
```

## Feature selection - SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

def Scale_Kbest(X,y,scaler):
    scaled_df=pd.DataFrame(scaler.fit_transform(X),columns=[X.columns])
    
    kbest=SelectKBest(score_func=f_regression,k=10)
    result=kbest.fit(X,y)
    dfcolumns=pd.DataFrame(X.columns)
    dfscores=pd.DataFrame(result.scores_)
    
    featureScores=pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns=['Feature','Score']
    
    return(featureScores.nlargest(10,'Score'))
    
```

### Function that implement Gridsearch decisionTree parameters with different K-folds cross validation, returns dataframe of best parameter and best score of each K-fold.

```python
def decision_tree_GridsearchCV(X,y):
    
    result_df=pd.DataFrame(index=range(0,6),columns=['K-Fold','Best parameter','Best score'])

    #  Set param_grid of GridSearchCV
    params = {
        'max_depth': range(2,11),
        'min_samples_split': (2,11),
        'criterion':['entropy','gini']
        }

    dt = DecisionTreeClassifier()
    # Gridsearch with different K-Fold cross validation, k= (5 to10)
    for i in range(5,11):    
        tree_gscv = GridSearchCV(dt, param_grid=params, cv=i, refit=True)
        tree_gscv.fit(X, y)
        result_df.iloc[i-5,:]=[i,tree_gscv.best_params_,tree_gscv.best_score_]
    return result_df    


```

## Classification - KNN

### Function that implement Scaling and Gridsearch KNN parameters with different K-folds cross validation, which returns dataframe of best parameter and best score of each K-fold.

```python
from sklearn.neighbors import KNeighborsClassifier


def Scale_Knn_gridsearch(scaler,X,y):
    
    result_df=pd.DataFrame(index=range(0,6),columns=['K-Fold','Best parameter','Best score'])
    
    X=pd.DataFrame(scaler.fit_transform(X),columns=[X.columns])
   
    # KNN gridsearch parameter setting
    param_grid={'n_neighbors':np.arange(1,21)}
    
    knn=KNeighborsClassifier()
    
    # Gridsearch with different K-Fold cross validation, k= (5 to10)
    for i in range(5,11):
        knn_gscv=GridSearchCV(knn,param_grid,cv=i)
        knn_gscv.fit(X,y)
        result_df.iloc[i-5,:]=[i,knn_gscv.best_params_,knn_gscv.best_score_]
    
    return (result_df)
```

### Function call with different scalers

```python
Scale_Knn_gridsearch(MinMaxScaler(),X,y)
```

```python
Scale_Knn_gridsearch(StandardScaler(),X,y)
```

```python
Scale_Knn_gridsearch(RobustScaler(),X,y)
```

## PCA into component =2

```python
from sklearn.decomposition import PCA
#PCA with n=2
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(mmDf.values)
principalDF=pd.DataFrame(data=principalComponents,columns=['principal component1','principal component2'])
print(principalDF)
```

```python
# Check the inertias(value of cohesion) to find the optimal number of clusterings

# List for storing inertias results
inertias=[]
ks=range(2,16)
for k in ks:
    kmeans=KMeans(n_clusters=k).fit(principalDF)
    # Save inertias in list
    inertias.append(kmeans.inertia_)
    cluster=kmeans.labels_
    print('Num of cluster: {}'.format(k), 'Inertias : {}'.format(kmeans.inertia_),"Silhouette Score",silhouette_score(principalDF,cluster))
    
    
# Plot inertias by cluster number
plt.figure(figsize=(15,6))
plt.plot(ks,inertias,'-o')
plt.xlabel('num of cluster')
plt.ylabel('inertias')
plt.xticks(ks)
plt.show()
```
